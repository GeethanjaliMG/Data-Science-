```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data=pd.read_csv('data.csv',names=["col1", "Col2"])
data.head(5)
```

```python
col1 = data.iloc[:,0]
col2 = data.iloc[:,1]
plt.scatter(col1,col2)
plt.show()
```

```python
#Step 1: initializing the values
m = 0
c = 0
```

```python
#Step 2: Initialize the learning Rate and Iterations
L = 0.0001
iter_ = 1000

n = float(len(col1))
print(n)

#For loop - Performing Gradient Descent 
for i in range(iter_):
    col2_pred = (m*col1)+c #The current prediction value 
    Diff_m = (-2/n)*sum(col1 * (col2 - col2_pred))
    Diff_c = (-2/n)*sum(col2 - col2_pred)
    m = m - (L*Diff_m) 
    c = c - (L*Diff_c)
    print("The Slope and intercept for iteration:",i+1)
    print("Slope: ",m)
    print("Intercept: ",c)

```
